```python
import pandas as pd
import re
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from openai import OpenAI

# 1. 환경 설정
# 보안 주의: 이 토큰은 반드시 삭제하고 새로 발급받으세요!
token = "YOUR_GITHUB_TOKEN" 

client = OpenAI(
    base_url="https://models.inference.ai.azure.com",
    api_key=token,
)

# 2. 데이터 로드
def load_data(file_path):
    try:
        return pd.read_csv(file_path, encoding='utf-8')
    except:
        return pd.read_csv(file_path, encoding='cp949')

df = load_data('french.csv')

# 3. 유연한 검색 로직
def find_relevant_data(user_query, dataframe):
    query = re.sub(r'[?.!, ]', '', user_query)
    for _, row in dataframe.iterrows():
        db_korean = re.sub(r'[?.!, ]', '', str(row['korean']))
        if query in db_korean or db_korean in query:
            return row
    return None

# 4. LLM 해설 생성 (구조화된 답변 생성)
def get_tutor_response(matched_row):
    system_prompt = """당신은 'NEW LEARN' 플랫폼의 왕초보 전용 프랑스어 선생님입니다.
    사용자에게 다음의 '초개인화 해설'을 제공하세요:
    
    1. [문법 조립 블록]: 문장을 단어별로 쪼개고 각각의 뜻을 설명하세요. (예: Je(나) + m'appelle(이름이 ~이다))
    2. [입문자 눈높이 해설]: 발음 팁과 실제 파리 여행에서 이 문장이 쓰이는 상황을 다정하게 설명하세요.
    
    답변은 한국어로 작성하며, 프랑스어 발음은 한글로 병기합니다. 
    형식은 아주 깔끔하게 문단으로 나누어 대답해주세요."""

    user_prompt = f"""
    [데이터베이스 정보]
    - 문장: {matched_row['french']}
    - 뜻: {matched_row['korean']}
    - 발음: {matched_row['pronunciation']}
    """

    response = client.chat.completions.create(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
        model="gpt-4o", 
        temperature=0.3,
        max_tokens=1000,
    )
    return response.choices[0].message.content

# 5. [중요] Streamlit에서 호출할 핵심 함수
def get_chatbot_result(question):
    """
    사용자 질문을 받아 (해설텍스트, 이미지URL) 튜플을 반환합니다.
    """
    matched_data = find_relevant_data(question, df)
    
    if matched_data is not None:
        # 데이터가 있을 때: 해설 생성 후 이미지 URL과 함께 반환
        explanation = get_tutor_response(matched_data)
        image_url = matched_data['image_url']
        return explanation, image_url
    else:
        # 데이터가 없을 때: 예외 메시지와 None 반환
        fallback_msg = "죄송합니다. 데이터베이스에 없는 내용입니다. 인사나 식당 관련 질문을 해주세요."
        return fallback_msg, None

# --- 테스트 실행부 ---
# 나중에 Streamlit에서는 아래와 같은 방식으로 사용하게 됩니다.
ans_text, ans_image = get_chatbot_result("안녕하세요")

print("=== 최종 반환 값 확인 ===")
print(f"텍스트 결과:\n{ans_text}")
print(f"\n이미지 URL 결과: {ans_image}")
```
