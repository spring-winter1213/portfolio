```python
import pandas as pd
import re
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from openai import OpenAI

# 1. 데이터 로드
def load_data(file_path):
    try:
        return pd.read_csv(file_path, encoding='utf-8')
    except:
        return pd.read_csv(file_path, encoding='cp949')

df = load_data('french.csv')

# 2. 검색 로직
def find_relevant_data(user_query, dataframe):
    if dataframe is None: return None
    query = re.sub(r'[?.!, ]', '', str(user_query)).lower()
    for _, row in dataframe.iterrows():
        if 'korean' in row and pd.notna(row['korean']):
            db_korean = re.sub(r'[?.!, ]', '', str(row['korean'])).lower()
            if query in db_korean or db_korean in query:
                return row
    return None

# 3. 테스트 입력 세트 (DB에 있는 것 1개 + DB에 없는 자유 질문 2개)
test_inputs = [
    "안녕하세요", 
    "프랑스어로 예의 바르게 거절하는 법 알려줘.",
    "파리 식당에서 팁을 줘야 하는지 문법적 표현과 함께 설명해줄래?"
]
```

```python
# 보안 주의: 실행 직전에 본인의 진짜 토큰으로 교체하세요.
token = "YOUR_GITHUB_TOKEN" 
openai_client = OpenAI(
    base_url="https://models.inference.ai.azure.com",
    api_key=token,
)

# 파라미터에 top_p가 추가되었습니다. (기본값 1.0)
def get_openai_response(question, temperature=0.3, top_p=1.0):
    matched_data = find_relevant_data(question, df)
    
    system_prompt = """당신은 'NEW LEARN' 플랫폼의 왕초보 전용 프랑스어 선생님입니다.
    사용자에게 다음의 '초개인화 해설'을 제공하세요:
    
    1. [문법 조립 블록]: 문장을 단어별로 쪼개고 각각의 뜻을 설명하세요. (예: Je(나) + m'appelle(이름이 ~이다))
    2. [입문자 눈높이 해설]: 발음 팁과 실제 파리 여행에서 이 문장이 쓰이는 상황을 다정하게 설명하세요.
    
    답변은 한국어로 작성하며, 프랑스어 발음은 한글로 병기합니다. 
    형식은 아주 깔끔하게 문단으로 나누어 대답해주세요."""

    if matched_data is not None:
        user_prompt = f"""
        [데이터베이스 정보]
        - 문장: {matched_data['french']}
        - 뜻: {matched_data['korean']}
        - 발음: {matched_data['pronunciation']}
        """
    else:
        user_prompt = f"데이터베이스에 없는 내용입니다. 선생님의 지식을 바탕으로 다음 질문에 친절하게 답변해 주세요: {question}"

    try:
        response = openai_client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt},
            ],
            model="gpt-4o", 
            temperature=temperature,
            top_p=top_p, # 복합 튜닝을 위해 추가됨
            max_tokens=1000,
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"OpenAI Error: {e}"
```

```python
# EXAONE 모델 로드 (이미 로드되어 있다면 이 부분은 주석 처리해도 무방합니다)
MODEL_NAME = "LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct"
print("EXAONE 모델 로딩 중...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)
print("EXAONE 로딩 완료!")

# 파라미터에 top_p가 추가되었습니다.
def get_exaone_response(question, temperature=0.3, top_p=1.0):
    matched_data = find_relevant_data(question, df)
    
    system_prompt = """당신은 'NEW LEARN' 플랫폼의 왕초보 전용 프랑스어 선생님입니다.
    사용자에게 다음의 '초개인화 해설'을 제공하세요:
    
    1. [문법 조립 블록]: 문장을 단어별로 쪼개고 각각의 뜻을 설명하세요. (예: Je(나) + m'appelle(이름이 ~이다))
    2. [입문자 눈높이 해설]: 발음 팁과 실제 파리 여행에서 이 문장이 쓰이는 상황을 다정하게 설명하세요.
    
    답변은 한국어로 작성하며, 프랑스어 발음은 한글로 병기합니다. 
    형식은 아주 깔끔하게 문단으로 나누어 대답해주세요."""

    if matched_data is not None:
        user_prompt = f"""
        [데이터베이스 정보]
        - 문장: {matched_data['french']}
        - 뜻: {matched_data['korean']}
        - 발음: {matched_data['pronunciation']}
        """
    else:
        user_prompt = f"데이터베이스에 없는 내용입니다. 선생님의 지식을 바탕으로 다음 질문에 친절하게 답변해 주세요: {question}"

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    
    prompt_str = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt_str, return_tensors="pt").to(model.device)
    
    output = model.generate(
        inputs.input_ids,
        max_new_tokens=1000,
        temperature=temperature,
        top_p=top_p, # 복합 튜닝을 위해 추가됨
        do_sample=True,
        eos_token_id=tokenizer.eos_token_id
    )
    
    response_text = tokenizer.decode(output[0][inputs.input_ids.shape[-1]:], skip_special_tokens=True)
    return response_text.strip()
```

- **비용 효율적 라우팅(Routing):** API 호출 비용 절감을 위해, 단순 DB 검색으로 해결 가능한 인사말/단어(CSV 내재 데이터)는 가벼운 로컬 모델(`EXAONE 3.5`)이 포매팅을 전담하고, 사용자가 DB에 없는 복잡한 문맥이나 문화적 설명을 요구할 때만 튜닝된 `GPT-4o`로 프롬프트를 넘기는 **멀티 모델 라우터 시스템**을 추후 구현하여 효율성을 높일 수 있음.

```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer, util
from openai import OpenAI

# 1. 임베딩 모델 로드 (한국어 의미 분석 특화 모델)
print("Sentence-Transformer 임베딩 모델 로딩 중...")
embedder = SentenceTransformer('jhgan/ko-sroberta-multitask')

# 2. 데이터 로드
def load_data(file_path):
    try:
        return pd.read_csv(file_path, encoding='utf-8')
    except:
        return pd.read_csv(file_path, encoding='cp949')

df = load_data('french.csv')
df['korean'] = df['korean'].fillna("") # 결측치 처리

# 3. 데이터베이스 전체를 미리 벡터(숫자)로 변환
corpus = df['korean'].tolist()
corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True)
print("✅ CSV 데이터베이스 벡터화 완료!")

# [비교용] 기존의 단순 키워드 검색 함수
def find_relevant_data_keyword(user_query):
    import re
    query = re.sub(r'[?.!, ]', '', str(user_query)).lower()
    for _, row in df.iterrows():
        db_korean = re.sub(r'[?.!, ]', '', str(row['korean'])).lower()
        if query in db_korean or db_korean in query:
            return row
    return None

# [신규] Sentence-Transformers 기반 의미 검색 함수
def find_relevant_data_vector(user_query, threshold=0.5):
    # 사용자의 질문을 벡터로 변환
    query_embedding = embedder.encode(user_query, convert_to_tensor=True)
    
    # DB 벡터들과 코사인 유사도(Cosine Similarity) 계산
    cos_scores = util.cos_sim(query_embedding, corpus_embeddings)[0]
    
    # 가장 유사도가 높은 항목 추출
    best_score = torch.max(cos_scores).item()
    best_idx = torch.argmax(cos_scores).item()
    
    # 유사도가 기준치(threshold) 이상일 때만 반환
    if best_score >= threshold:
        return df.iloc[best_idx], best_score
    else:
        return None, best_score
```

```python
# 보안 주의: 실행 직전에 본인의 진짜 토큰으로 교체하세요.
token = "YOUR_GITHUB_TOKEN" 
openai_client = OpenAI(
    base_url="https://models.inference.ai.azure.com",
    api_key=token,
)

def get_tutor_response(question, search_method="vector", temperature=0.5, top_p=0.8):
    # 1. 사용자가 선택한 방식에 따라 검색 수행
    if search_method == "keyword":
        matched_data = find_relevant_data_keyword(question)
        score_info = "(단순 키워드 매칭)"
    else:
        matched_data, score = find_relevant_data_vector(question)
        score_info = f"(유사도: {score:.2f})" if matched_data is not None else ""

    # 2. 프롬프트 구성
    system_prompt = """당신은 'NEW LEARN' 플랫폼의 왕초보 전용 프랑스어 선생님입니다.
    1. [문법 조립 블록]: 문장을 단어별로 쪼개고 각각의 뜻을 설명하세요.
    2. [입문자 눈높이 해설]: 발음 팁과 실제 파리 여행에서 쓰이는 상황을 다정하게 설명하세요.
    답변은 한국어로 작성하며, 프랑스어 발음은 한글로 병기합니다."""

    if matched_data is not None:
        user_prompt = f"""
        [데이터베이스에서 검색된 정보 {score_info}]
        - 문장: {matched_data['french']}
        - 뜻: {matched_data['korean']}
        - 발음: {matched_data['pronunciation']}
        
        위 정보를 바탕으로 사용자에게 친절하게 해설해 주세요.
        """
    else:
        user_prompt = f"데이터베이스에 없는 내용입니다. 선생님의 지식을 바탕으로 다음 질문에 답변해 주세요: {question}"

    # 3. LLM 호출
    try:
        response = openai_client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt},
            ],
            model="gpt-4o", 
            temperature=temperature,
            max_tokens=800,
        )
        # 결과 텍스트 맨 앞에 어떤 방식으로 검색했는지 꼬리표를 달아줌
        prefix = f"🔍 [검색 방식: {search_method.upper()}] 매칭 여부: {'성공' if matched_data is not None else '실패'}\n\n"
        return prefix + response.choices[0].message.content.strip()
    except Exception as e:
        return f"OpenAI Error: {e}"
```

```python
print("="*80)
print("🚀 [과제수행 1] 자연어 처리 기법 도입 전/후 검색 정확도 비교")
print("="*80)

# 테스트 질문: "아침 인사" 
# (만약 CSV에 '안녕하세요'라고만 되어있다면 키워드 검색은 무조건 실패함)
test_q = "아침에 카페에서 쓸 수 있는 인사말 알려줘"

print(f"\n▶ 질문: '{test_q}'\n")

print("🔴 [기존 방식: 단순 문자열 키워드 검색 + GPT]")
print("─" * 80)
print(get_tutor_response(test_q, search_method="keyword"))

print("\n\n🔵 [개선 방식: Sentence-Transformers 의미 검색 + GPT]")
print("─" * 80)
print(get_tutor_response(test_q, search_method="vector"))
```

## 📊 프로젝트 필수 수행: 최종 결과 분석 및 개선 시도

### 1. NLP 추가 모델(Sentence-Transformers) 도입 및 결과 비교
- **문제 정의:** 기존에는 단순 정규식(Regex)을 이용한 키워드 매칭 로직을 사용함. 이 경우 사용자가 "아침 인사"라고 물었을 때, CSV의 `korean` 열에 정확히 "아침 인사"라는 글자가 없으면 매칭에 실패하고 LLM 자유 추론으로 넘어가버리는(할루시네이션 위험 증가) 한계가 있었음.
- **개선 시도:** 한국어 의미 분석에 특화된 Sentence-Transformers(`jhgan/ko-sroberta-multitask`) 임베딩 모델을 추가로 도입하여 **Vector Search(코사인 유사도 기반 의미 검색)** 파이프라인을 구축함.
- **결과 분석:** 도입 후, 사용자가 "아침에 카페에서 쓸 수 있는 인사말" 등 뭉뚱그려 질문해도 임베딩 공간 상에서 가장 의미가 가까운 "안녕하세요(Bonjour)" 데이터를 정확히 찾아내어 LLM에게 전달함. RAG(검색 증강 생성) 아키텍처의 검색 정확도(Retrieval Accuracy)가 비약적으로 향상됨.

### 3. 최종 아키텍처 및 개선 방향
- **최종 아키텍처:** `Sentence-Transformers(의미 검색)` + `OpenAI GPT-4o(Temperature 0.3)`
- **결론:** 단순히 프롬프트만 LLM에 던지는 1차원적인 챗봇에서 벗어나, NLP 임베딩 모델로 자체 데이터베이스의 검색 성능을 극대화하고, LLM이 이를 바탕으로 세련된 튜팅 템플릿을 생성하는 완전한 형태의 AI 튜터 서비스를 구현하였음. 향후 데이터베이스가 수만 건으로 늘어날 경우 FAISS 등 벡터 DB를 도입하면 검색 속도를 더욱 최적화할 수 있을 것으로 기대됨.

```python
# 보안 주의: 실행 직전에 본인의 진짜 토큰으로 교체하세요.
token = "YOUR_GITHUB_TOKEN"  
openai_client = OpenAI(
    base_url="https://models.inference.ai.azure.com",
    api_key=token,
)

# 튜닝에서 찾은 최적값을 기본값으로 설정합니다.
def get_tutor_response(question, search_method="vector", temperature=0.5, top_p=0.8):
    # 1. 사용자가 선택한 방식에 따라 검색 수행
    if search_method == "keyword":
        matched_data = find_relevant_data_keyword(question)
        score_info = "(단순 키워드 매칭)"
    else:
        # 유사도 점수와 함께 데이터를 가져옵니다 (1번 분석용 로직 적용)
        matched_data, score = find_relevant_data_vector(question)
        score_info = f"(유사도: {score:.2f})" if matched_data is not None else ""

    # 2. 프롬프트 구성 (사용자 원본 프롬프트 유지)
    system_prompt = """당신은 'NEW LEARN' 플랫폼의 왕초보 전용 프랑스어 선생님입니다.
    1. [문법 조립 블록]: 문장을 단어별로 쪼개고 각각의 뜻을 설명하세요.
    2. [입문자 눈높이 해설]: 발음 팁과 실제 파리 여행에서 이 문장이 쓰이는 상황을 다정하게 설명하세요.
    답변은 한국어로 작성하며, 프랑스어 발음은 한글로 병기합니다."""

    if matched_data is not None:
        user_prompt = f"""
        [데이터베이스에서 검색된 정보 {score_info}]
        - 문장: {matched_data['french']}
        - 뜻: {matched_data['korean']}
        - 발음: {matched_data['pronunciation']}
        
        위 정보를 바탕으로 사용자에게 친절하게 해설해 주세요.
        """
    else:
        # DB에 없을 때 스스로 추론하는 로직
        user_prompt = f"데이터베이스에 없는 내용입니다. 선생님의 지식을 바탕으로 다음 질문에 답변해 주세요: {question}"

    # 3. LLM 호출 (최적 파라미터 0.5 / 0.8 적용)
    try:
        response = openai_client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt},
            ],
            model="gpt-4o", 
            temperature=temperature,
            top_p=top_p,
            max_tokens=1000,
        )
        
        # 어떤 방식으로 검색했는지 결과를 명시함
        status = "성공" if matched_data is not None else "실패"
        prefix = f"🔍 [방식: {search_method.upper()} | 매칭: {status} {score_info}]\n"
        return prefix + "-"*50 + "\n" + response.choices[0].message.content.strip()
    except Exception as e:
        return f"OpenAI Error: {e}"
```

```python
print("="*80)
print("🚀 [과제수행 1] 자연어 처리 기법(Vector Search) 도입 전/후 검색 정확도 비교")
print("="*80)

# 테스트 질문: "아침 인사" 
# (CSV에 '안녕하세요'라고만 되어 있는 경우를 가정하여 의미 검색의 우수성을 테스트)
test_q = "아침에 카페에서 점원에게 할 수 있는 인사 알려줘"

print(f"▶ 테스트 질문: '{test_q}'")
print("▶ 설정 파라미터: Temperature=0.5, Top_p=0.8 (최적값 적용)\n")

print("🔴 [과거의 나: 단순 키워드 검색 + GPT]")
print("-" * 80)
# '아침', '카페', '점원'이라는 단어가 CSV에 없으면 매칭 실패함
print(get_tutor_response(test_q, search_method="keyword"))

print("\n\n🔵 [현재의 나: 벡터 임베딩 의미 검색 + GPT]")
print("-" * 80)
# 의미상 가장 유사한 '안녕하세요(Bonjour)'를 찾아내어 RAG 기반 답변을 생성함
print(get_tutor_response(test_q, search_method="vector"))
```
