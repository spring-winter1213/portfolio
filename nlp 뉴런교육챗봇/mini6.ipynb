```python
# %%
# =========================================
# [Step 0] 라이브러리
# =========================================
import pandas as pd
import torch
import re
import matplotlib.pyplot as plt
import numpy as np
from sentence_transformers import SentenceTransformer, util
from openai import OpenAI

# %%
# =========================================
# [Step 1] 데이터 로드
# =========================================
def load_data(file_path):
    try:
        return pd.read_csv(file_path, encoding='utf-8')
    except:
        return pd.read_csv(file_path, encoding='cp949')

df = load_data('french.csv')
df['korean'] = df['korean'].fillna("")

# %%
# =========================================
# [Step 2] 키워드 검색 (Baseline)
# =========================================
def find_keyword(query):
    query = re.sub(r'[?.!, ]', '', str(query)).lower()
    
    for _, row in df.iterrows():
        db = re.sub(r'[?.!, ]', '', str(row['korean'])).lower()
        if query in db or db in query:
            return row
    return None

# %%
# =========================================
# [Step 3] 임베딩 기반 RAG
# =========================================
print("🔄 임베딩 모델 로딩 중...")
embedder = SentenceTransformer('jhgan/ko-sroberta-multitask')

corpus = df['korean'].tolist()
corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True)
print("✅ 벡터화 완료!")

def find_vector(query, threshold=0.5):
    query_embedding = embedder.encode(query, convert_to_tensor=True)
    scores = util.cos_sim(query_embedding, corpus_embeddings)[0]

    best_idx = torch.argmax(scores).item()
    best_score = scores[best_idx].item()

    if best_score >= threshold:
        return df.iloc[best_idx], best_score
    return None, best_score

# %%
# =========================================
# [Step 4] OpenAI 설정
# =========================================
client = OpenAI(api_key="YOUR_OPENAI_API_KEY")  # 🔑 API 키 입력

def generate(prompt, temp=0.5, top_p=0.8):
    res = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role":"user","content":prompt}],
        temperature=temp,
        top_p=top_p,
        max_tokens=500
    )
    return res.choices[0].message.content.strip()

# %%
# =========================================
# [Step 5] RAG 챗봇
# =========================================
def rag_chat(query, temp=0.5, top_p=0.8):
    data, score = find_vector(query)

    if data is not None:
        prompt = f"""
문장: {data['french']}
뜻: {data['korean']}
발음: {data['pronunciation']}

초보자에게 설명해줘
"""
    else:
        prompt = query

    return generate(prompt, temp, top_p)

# %%
# =========================================
# [Step 6] 키워드 vs RAG 성능 비교
# =========================================
tests = ["안녕하세요", "카페 인사", "아침 인사"]

kw, vec = 0, 0

for t in tests:
    print("\n질문:", t)

    # 키워드
    k = find_keyword(t)
    if k is not None:
        print("🔴 키워드: 성공")
        kw += 1
    else:
        print("🔴 키워드: 실패")

    # RAG
    v, score = find_vector(t)
    if v is not None:
        print(f"🔵 RAG: 성공 (유사도 {score:.2f})")
        vec += 1
    else:
        print("🔵 RAG: 실패")

# 성공률 계산
kw_rate = kw / len(tests) * 100
vec_rate = vec / len(tests) * 100

print("\n📊 성능 결과")
print(f"키워드 검색: {kw_rate:.1f}%")
print(f"RAG 검색: {vec_rate:.1f}%")

# %%
# =========================================
# [Step 7] 성능 그래프
# =========================================
labels = ['Keyword', 'RAG']
values = [kw_rate, vec_rate]

plt.figure()
plt.bar(labels, values)

plt.title('검색 성능 비교')
plt.ylabel('성공률 (%)')

for i, v in enumerate(values):
    plt.text(i, v + 2, f"{v:.1f}%", ha='center')

plt.ylim(0, 110)
plt.show()

# %%
# =========================================
# [Step 8] 🔥 파라미터 튜닝 평가
# =========================================
temps = [0.1, 0.5, 1.0]
lengths = []
variances = []

test_q = "프랑스어 사랑 표현 알려줘"

for t in temps:
    outputs = [rag_chat(test_q, temp=t) for _ in range(3)]

    lens = [len(o) for o in outputs]

    lengths.append(np.mean(lens))
    variances.append(np.var(lens))

print("\n📊 튜닝 결과")
print("Temperature:", temps)
print("평균 길이:", lengths)
print("분산:", variances)

# %%
# =========================================
# [Step 9] 튜닝 그래프
# =========================================
plt.figure()
plt.plot(temps, lengths, marker='o')
plt.title('Temperature vs 응답 길이')
plt.xlabel('Temperature')
plt.ylabel('길이')
plt.show()

plt.figure()
plt.plot(temps, variances, marker='o')
plt.title('Temperature vs 다양성')
plt.xlabel('Temperature')
plt.ylabel('분산')
plt.show()

# %%
# =========================================
# [Step 10] 슬라이드용 출력
# =========================================
sentence = "Je t'aime plus que toutes les étoiles"

for t in temps:
    print(f"\n🔥 Temp={t}")
    print(rag_chat(sentence, temp=t))

# %%
# =========================================
# [Step 11] 최종 챗봇 테스트
# =========================================
print("\n🚀 최종 챗봇 테스트")

for t in tests:
    print("\n질문:", t)
    print(rag_chat(t))
```
